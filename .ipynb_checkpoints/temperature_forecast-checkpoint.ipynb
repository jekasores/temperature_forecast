- There are many different dates to be displayed in a bar plot. To be exact there are 307 different dates
- It might be interesting to verify from what year is this data. We can also think about questions as:
    - Is there data available from which years?
    - Is there data for every month of the year?
    - Is there data for every day in a month?

```python
#Checking the date format to see how we are going to separate year, month, and day info
df.Date[0]
```

```python
df.Date[0][:4] #year
```

```python
df.Date[0][5:7] #month
```

```python
df.Date[0][8:] #day
```

```python
df["year"] = df["Date"].apply(lambda x: x[:4]) #get the year from the date
df["month"] = df["Date"].apply(lambda x: x[5:7]) #get the month from the date
df["day"] = df["Date"].apply(lambda x: x[8:]) #get the day from the date
```

```python
df[["year", "month", "day"]]
```

**Is there data available from which years?**
    - There are data from 2013, 2014, 2015, 2016, and 2017

```python
df.year.value_counts()
```

```python
def check_data_month(df, year):
    df = df[df.year==year]
    print("Checking if there is only data from the selected year: ")
    print(df.year.value_counts())
    print("Checking the months available: ")
    print(df.month.value_counts())
```

```python
years = ["2013", "2014", "2015", "2016", "2017"]
for y in years:
    check_data_month(df, y)
    print("______________ \n")
```

**Is there data for every month of the year?**
    - For every year, there is just data available for June, July, and August
**Is there data for every day in a month?**
    - No
