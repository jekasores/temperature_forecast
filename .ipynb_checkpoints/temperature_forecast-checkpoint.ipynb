## 4) Machine learnig model  <a class="anchor" id="ml-model"></a>
- **Objective:** Forecast next-day maximum and minimum air temperatures in Seoul, South Korea
    - Temperature forecasts enable authorits to act to mitigate the damage caused by heat waves and tropical nights
- **Method:** Linear regression
    - The [paper](https://agupubs.onlinelibrary.wiley.com/doi/epdf/10.1029/2019EA000740) uses Random forest (RF), support vector regression (SVR), artificial network (ANN), and multi-model ensemble (MME) to solver the problem. It also uses the R squared, RMSE, and bias to evaluate the temperature prediction.
    - To simplify this analysis aming to have an initial result fast, I chose linear regression to predict the temperature and R squared to evaluate the model.

```python
#Droping unecessary columns
df.drop(columns= ["station", "Date"], inplace=True)
df.columns
```

```python
#Transforming date columns from string to int
columns_date = ["year", "month", "day"]
for col in columns_date:
    df = string_into_int(df, col)
```

```python
#Split data in X and y
X = X = df.drop(["Next_Tmax", "Next_Tmin"], axis='columns')
y_max = df["Next_Tmax"]
y_min = df["Next_Tmin"]

print(X.shape)
print(y_max.shape)
print(y_min.shape)
```

```python
#Split into train and test
X_train_max, X_test_max, y_train_max, y_test_max = train_test_split(X, y_max, test_size=.30, random_state=42)
X_train_min, X_test_min, y_train_min, y_test_min = train_test_split(X, y_min, test_size=.30, random_state=42)
```

```python
#Instantiate model
lm_model_max = LinearRegression(normalize=True)
lm_model_min = LinearRegression(normalize=True)
```

```python
#Fit
lm_model_max.fit(X_train_max, y_train_max)
lm_model_min.fit(X_train_min, y_train_min)
```

```python
#Predict and score the model
y_test_preds_max = lm_model_max.predict(X_test_max)
y_test_preds_min = lm_model_min.predict(X_test_min)

rsquared_score_max = r2_score(y_test_max, y_test_preds_max)
rsquared_score_min = r2_score(y_test_min, y_test_preds_min)


#Rsquared and y_test
length_y_test_max = len(y_test_preds_max)
length_y_test_min = len(y_test_preds_min)
print("The r-squared score for your model_max was {} on {} values.".format(rsquared_score_max, length_y_test_max))
print("The r-squared score for your model_min was {} on {} values.".format(rsquared_score_min, length_y_test_min))
```

- R-squared varies beteween 0 and 1. The higher the value of r-squared, the more adjustable my model is to the data samples
- Two models were build to predict the maximum (Next_Tmax) and minimum temperature (Next_Tmin) called lm_model_max and lm_model_min respectvely. Both models show good performance for the test data.
- More theorical information about r-square can be found [here](https://pt.wikipedia.org/wiki/Coeficiente_de_determina%C3%A7%C3%A3o)

```python
#To check which features matter in the model, check the weight of the coeficients
#Because the features are normalized we can look at how large the coeficient is
coef_df_max = coef_weights(lm_model_max.coef_, X_train_max, lm_model_max)
coef_df_max.head(20)
```

```python
coef_df_min = coef_weights(lm_model_min.coef_, X_train_min, lm_model_min)
coef_df_min.head(20)
```

- When we analyze the coeficients of each feature (absolute value) we can identify each ones have the highest importance for the model prediction
- Both Next_Tmax and Next_Tmin have similar main features that impact in temperatura. These features are: latitude (lat), month of the year (month), cloud (LDAPS variables), longitude (lon), and slope.
